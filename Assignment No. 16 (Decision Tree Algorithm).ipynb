## 2. What are the Steps for Making a decision tree?


**1 Select the Root Node :**
* Choose the attribute that best splits the dataset. This is typically done by evaluating a measure of Gini impurity, information gain(Classification) or  MSE, MAE reduction (for regression) for each attribute.

    - Classification : High IG or Low Gini Index
    - Regression     : Low MSE, MAE
        
**2.Split the Dataset:**
* Divide the dataset into subsets based on the chosen attribute. Each subset corresponds to a branch from the root node.

**3. Recursive Splitting:**
* Repeat the splitting process for each subset. Choose the best attribute at each internal node and split the data accordingly.

**4. Stopping Criteria:**
* Define conditions for when to stop splitting. This helps prevent overfitting.

   -->Common stopping criteria include:
   
            1.Maximum depth of the tree.
            2.Minimum number of samples in a node.
            3.Minimum impurity threshold.
            4.No further improvement in impurity reduction.
                
**5. Create Leaf Nodes:**
* When a stopping criterion is met, create a leaf node. Assign it the most common class label (for classification) or the mean value (for regression) of the samples in that node.

**6. Pruning (Optional):**
* After the tree is built, prune unnecessary branches to avoid overfitting. Pruning involves removing nodes that do not significantly contribute to improving the model's performance on unseen data.
        
**7. Final Tree:**
* The resulting structure is the decision tree. It consists of decision nodes, leaf nodes, and edges representing the decision-making process.
        
**8. Prediction:**
* To make a prediction for a new instance, traverse the tree from the root to a leaf node based on the attribute values of the instance. Return the class label (for classification) or numerical value (for regression) associated with the leaf node.
        
**9. Evaluate the Tree:**
* Assess the performance of the decision tree using metrics such as accuracy, precision, recall (for classification), or mean squared error (for regression) on a separate validation or test dataset.

**10. Fine-Tuning (Optional):**
* Adjust hyperparameters, such as the maximum depth, minimum samples per leaf, or impurity threshold, to optimize the decision tree's performance.

**Ensemble Methods:**
* Consider using ensemble methods like Random Forests or Gradient Boosted Trees for improved performance and robustness.


*Remember that decision trees are prone to overfitting, so it's crucial to apply appropriate stopping criteria and, if needed, pruning techniques to achieve a well-generalized model.*

## 5. Explain Decision tree Key terms: Root Node, Decision Node/Branch Node, Leaf or Terminal Node.

* **Root Node:** The initial node at the beginning of a decision tree, where the entire dataset starts dividing based on various features or conditions.

* **Decision/Branch Nodes:** Nodes resulting from the splitting of root nodes are known as decision nodes. These nodes represent intermediate decisions or conditions within the tree.

* **Leaf/Terminal Nodes :**  Leaf node is defined as the output of decision nodes, but if they do not contain any branch, it means the tree cannot be segregated further from this node. Leaf nodes are also referred to as terminal nodes.

## 6. What are Assumptions while creating a Decision Tree

**Binary Splits:**
* Decision trees typically assume binary splits at each node. This means that at each decision point, the data is split into two branches based on a particular attribute or condition

**Attribute Independence:** 
* Decision trees assume that attributes used for splitting at each node are independent of each other.

**Impurity Measures:**
* Decision tree algorithms use impurity measures (e.g., Gini impurity, entropy) to determine the best splits. The assumption here is that minimizing impurity leads to better decision boundaries. 

**Recursive Partitioning:** 
* Decision trees are built through recursive partitioning, where the dataset is repeatedly split until a stopping criterion is met. 

**Biased Toward Dominant Classes:**
* In classification tasks with imbalanced class distribution, decision trees may be biased toward the dominant class. Techniques such as class weighting or resampling can help address this issue.

**Axis-Aligned Splits:**
* Consideration: Decision trees use axis-aligned splits, meaning that splits are made parallel to the coordinate axes. This can be a limitation when dealing with non-linear relationships, as decision trees may struggle to capture complex patterns that require diagonal or curved splits.

## 22. Explain Linear regression vs decision trees regression


*Linear regression and decision tree regression are both techniques used in machine learning for predicting numerical values*

**1. Principle**
* Linear regression is a linear approach to modeling the relationship between a dependent variable (target) and one or more independent variables (features or predictors). It assumes that the relationship between the variables can be represented by a linear equation
* Decision tree regression, on the other hand, is a non-linear model that makes predictions by partitioning the input space into regions and assigning a constant value to each region. It builds a tree-like structure where each internal node represents a decision based on a feature, and each leaf node represents the predicted value.

**2. Model Form :**
* Mathimatical equation of the linear regression is :
   y = m1x1 + m2x2 + c
* Instead of a mathematical equation, decision tree regression is represented by a tree structure, where each leaf node contains the predicted value for the corresponding input region.

**3. Training :**
* The model is trained by finding the values of the coefficients that minimize the sum of the squared differences between the predicted and actual values.
* The tree is constructed by recursively splitting the data based on features to minimize the variance of the predicted values within each region. The splitting process continues until a stopping criterion is met.

**4. Interpretability :**
* Linear regression models are often easy to interpret, and the coefficients provide insights into the relationship between variables.
* While decision trees can capture complex relationships, they can be harder to interpret compared to linear regression models.

The simple difference is :

**Linearity:** Linear regression assumes a linear relationship, while decision trees can capture non-linear patterns.

**Interpretability:** Linear regression models are often more interpretable, especially when dealing with a small number of features. Decision trees might be more complex to interpret, especially with deep trees.

**Handling Features:** Linear regression assumes a linear relationship with the features, while decision trees can handle non-linear relationships and interactions between features more naturally.

**Overfitting:** Decision trees are prone to overfitting, especially if the tree is deep. Techniques like pruning or using ensemble methods (e.g., Random Forest) can mitigate this issue.

**Robustness:** Linear regression can be sensitive to outliers, while decision trees are less affected by them.

## 24. What is max_depth in the Decision Tree?

*The max_depth hyperparameter in a decision tree controls the maximum depth of the tree. In other words, it limits the number of nodes between the root and the farthest leaf node.*

* **Definition:** max_depth is an integer parameter that determines the maximum depth of the decision tree. If this parameter is set to a positive integer, the tree will stop growing when it reaches the specified maximum depth.

* **Tuning:** By tuning the max_depth hyperparameter, you can control the size and complexity of the decision tree. A smaller max_depth value will result in a shallower tree, while a larger value will allow the tree to grow deeper.

* **Overfitting:** Setting a higher max_depth can lead to overfitting, where the model captures noise in the training data and performs poorly on new, unseen data. On the other hand, setting a lower max_depth may result in a simpler model that generalizes better to new data.

## 25. What are the min_samples_split and min_samples_leaf hyperparameters?

*In decision tree algorithms, the min_samples_split and min_samples_leaf hyperparameters are used to control the size of the nodes during the construction of the tree. These hyperparameters help prevent overfitting by setting criteria on the minimum number of samples required to split an internal node (min_samples_split) and the minimum number of samples required to form a leaf node (min_samples_leaf).*

**1. min_samples_split:**

* **Definition:** It represents the minimum number of samples required to split an internal node. If the number of samples at a node is less than min_samples_split, the node will not be split, and it will become a leaf node.
* **Tuning:** Higher values for min_samples_split result in a tree with fewer splits, promoting simpler models and potentially reducing overfitting. However, setting it too high might lead to underfitting.

**2. min_samples_leaf:**

* **Definition:** It represents the minimum number of samples required to be in a leaf node. If the number of samples in a node is less than min_samples_leaf, the node is not split, and it becomes a leaf.
* **Tuning:** Similar to min_samples_split, higher values for min_samples_leaf result in larger leaf nodes, which can lead to a more generalized model. Like min_samples_split, setting it too high might lead to underfitting.

*Tuning these hyperparameters involves finding a balance between model complexity and the ability to generalize well to new, unseen data.*
