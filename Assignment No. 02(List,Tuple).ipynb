## 1. What is a List?

- It is used to store multiple items in a single variable.
- It is an ordered and changeable collection of data objects
- Heterogenous Data Type :
     - We can add any data type in list 
     - Any data type > int, float, complex, str, list, tuple, set, frozenset, dict, numpy array


Key characteristics of lists in Python include:

1. **Mutable:** Lists are mutable, meaning you can modify, add, or remove elements after the list is created. 
2. **Ordered:** Lists maintain the order of elements as they are added.
3. **Heterogeneous Elements:** Elements in a list can be of different data types, including integers, floats, strings, or even other lists.
4. **Indexed:** Elements in a list are accessed using index values. Indexing starts from 0 for the first element.
5. **Duplicate** objects are allowed
6. List is enclosed by [].
7. List items are comma seperated.

## 2. What is Tupple?

In Python, a tuple is a collection of ordered and **immutable elements**. Tuples are similar to lists, but they have a few key differences:

**1. Immutable:** Once a tuple is created, its elements cannot be modified, added, or removed. This property makes tuples suitable for representing fixed collections of values.

**2. Ordered**: Like lists, tuples maintain the order of elements. The order is based on the sequence in which elements are defined.

**3. Enclosed in Parentheses:** Tuples are defined by enclosing elements in parentheses ().




- Exactly same as list except it is **immutable**
- Ordered
- Duplicates are allowed
- Heterogenous objects are allowed 
- Enclosed by parenthesis >>()
- Comma Seperated 
- Any Data Type >> int, float, complex, str, list, tuple, set, frozenset, dict, numpy array

*NOTE : Paranthesis are optional*

## 4. Python Program to find the largest element in the list

```python
#Method 1 : with built in function
list1 = [10,100,50,70,150,200,350]
print('Largest Element in the list is :', max(list1))
```

```python
#Method 1 : without built in function
my_list = [12, 45, 78, 23, 56, 89]

# Check if the list is not empty
if my_list:
    largest = my_list[0]

    # Iterate through the list to find the largest element
    for element in my_list:
        if element > largest:
            largest = element

    print(f"The largest element in the list is: {largest}")
else:
    print("The list is empty.")
```

# 16. WAP to create two lists, one containing all even numbers and the other containing all odd numbers between 0 to 151

# 19. From a list containing ints, strings, and floats, make three lists to store them separately.


```python
list1   =[10, 20,'python', 'data science', 1.68, 56, 5.67, 'machine learning', 89, 18.45, 98, 'deep learning']
ints    =[]
strings =[]
floats  =[]

for i in list1:
    if type(i)==int:
        ints.append(i)
    
    elif type(i)==str:
         strings.append(i)
            
    elif type(i)==float:
         floats.append(i)

print('Integers :', ints)
print('Strings :', strings)
print('floats :', floats)
```

## 27. How to flatten a list in python?

```python
nested_list1=[[1,2,3,4],['python',' data science', 'machine learning'], [1.5, 6.7, 9.4], [(2+3j), (7+4j)]]
flatten_list=[]

for i in nested_list1:
    for j in i:
        flatten_list.append(j)

print(flatten_list)
```

## 40. How to flatten a list in python

```python
list1 = [[1,2,3,4,],['yogesh', 'ramesh', 'mahesh', 'ganesh'], [10.25,20.65,30.89,40.68],[3+5j, 8+6j]]
flatten_list=[]
for element in list1:
    for sub_element in element:
        flatten_list.append(sub_element)
print('flatten List is :', flatten_list)
```
